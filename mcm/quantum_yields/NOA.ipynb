```julia
# load in spectrometer data
hr4000_df = CSV.File("../hr4000_wavelengths.txt") |> DataFrame ; 
```

# Model Evaluation

```julia
λ = hr4000_df.λ
Φ = zeros(size(λ))

idxs = [idx for idx ∈ 1:size(λ,1) if λ[idx] ≥ 245.0 && λ[idx] ≤ 340.0 ]
Φ[idxs] .= 0.9



p = plot(
    λ,
    Φ,
    xlabel="λ [nm]",
    ylabel="Φ",
    alpha=0.7,
    lw=3,
    label="",
    title="Quantum Yield for NOA → CH₃C(O)CH₂O + NO₂",
    titlefontsize=12,
)
```
