# Model Evaluation

```julia
λ = hr4000_df.λ
Φ = zeros(size(λ))

idxs = [idx for idx ∈ 1:size(λ,1) if λ[idx] ≥ 248.0 ]
Φ[idxs] .= 1.0



p = plot(
    λ,
    Φ,
    xlabel="λ [nm]",
    ylabel="Φ",
    alpha=0.7,
    lw=3,
    label="",
    title="Quantum Yield for CH₃ONO₂ → CH₃O + NO₂",
    titlefontsize=12,
)
```
